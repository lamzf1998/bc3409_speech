```python
timesteps = 3
```

```python
df_size = len(df)
```

```python
print(df_size)
```

```python
df_resize = int(df_size//3)
```

```python
print(df_resize)
```

```python
num_class = 1
```

```python
X = df.iloc[:,:-1] 
```

```python
print(X)
```

```python
df_dim = len(X.columns)
```

```python
print(df_dim)
```

```python
df_trunc_size = df_resize * timesteps
```

```python
print(df_trunc_size)
```

```python
Y = df.iloc[:,-1]
```

```python
print(Y)
```

```python
print(X,Y)
```

```python
X = X.values
Y = Y.values
```

```python
print(X,Y)
```

```python
X = X[:df_trunc_size, :]
```

```python
Y = Y[:df_trunc_size]
```

```python
X.shape
```

```python
Y.shape
```

```python
X_shaped = X.reshape(df_resize, timesteps, df_dim)
```

```python
X_shaped.shape
```

```python
Y_shaped = Y.reshape(df_resize,timesteps)
```

```python
Y_shaped.shape
```

```python
Y_shaped = Y_shaped[:,-1]
```

```python
Y_shaped.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_shaped, Y_shaped, test_size=0.2)
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense, Dropout, SimpleRNN, LSTM, GRU
```

```python
model = Sequential()
```

```python
model.add(LSTM(8, input_shape=(timesteps, df_dim)))
```

```python
model.add(Dropout(0.3))
```

```python
model.add(Dense(num_class, activation = "sigmoid"))
```

```python
model.summary()
```

```python
model.compile(loss = "binary_crossentropy", metrics=["accuracy"])
```

```python
model.fit(X_train,Y_train, batch_size =10, epochs = 5)
```

```python
model.evaluate(X_test,Y_test)
```

```python
pred = model.predict(X_test)
```

```python
print(pred)
```

```python
pred = pred > 0.3
```
